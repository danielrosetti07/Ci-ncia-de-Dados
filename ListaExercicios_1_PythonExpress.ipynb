```python
# 1) Elabore um código que imprima a mensagem *"Olá terráqueos!!"* na tela.

print("Olá terráqueo!!")
```

```python
#Elabore um código que peça um número e então mostre a mensagem: "O número informado foi [número]."
numero = input("Escreva algum número:")
print(f"O número informado foi: {numero}")
```

```python
#Elabore um código que peça dois números e imprima a soma.
numero1 = float (input("Número 1: "))
numero2 = float (input("Número 2: "))
resultado = numero1 + numero2
print(f"O resultado da soma foi: {resultado}")
```

```python
#Elabore um código que peça as 4 notas bimestrais e mostre a média.
nota1 = float (input("Nota 1:"))
nota2 = float (input("Nota 2:"))
nota3 = float (input("Nota 3:"))
nota4 = float (input("Nota 4:"))
resultado = (nota1 + nota2 + nota3 + nota4) /4
print(f"Nota bimestral: {resultado}")
```

```python
#Elabore um código que converta metros para centímetros.
metros = float (input("Metros:"))
resultado = metros*100
print(f"Centímetros: {resultado}")
```

```python
#Elabore um código que peça o raio de um círculo, calcule e mostre sua área.
raio = float (input("Raio do círculo:"))
resultado = 3.14*(raio**2)
print(f"Área do círculo: {resultado}")
```

```python
#Elabore um código que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário.
lado = float (input("Tamanho do lado:"))
resultado = lado*lado*2
print(f"Dobro da área do quadrado: {resultado}")
```

```python
#Elabore um código que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês em uma mensagem amigável.
salario = float (input("Quanto você ganha por hora:"))
hora = float (input("Número de horas trabalhado no mês:"))
resultado = salario*hora
print(f"Este é seu salário: {resultado}")
```

```python
#Elabore um código que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius. ** Onde C = 5 * ((F-32) / 9)**.
fahrenheit = float (input("Temperatura em Fahrenheit:"))
resultado =  5 * ((fahrenheit-32) / 9)
print(f"Temperatura em celsius: {resultado}")
```

```python
#Elabore um código que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.
celsius = float (input("Temperatura em celsius: "))
resultado = (celsius*9/5)+32
print(f"Temperatura em fahrenheit: {resultado}")
```

```python
#Elabore um código que peça 2 números inteiros e um número real. Calcule e mostre:

#o produto do dobro do primeiro com metade do segundo .
#a soma do triplo do primeiro com o terceiro.
#o terceiro elevado ao cubo.
Numero1 = int (input("Número 1: "))
Numero2 = int (input("Número 2: "))
Numero3 = float (input("Número 3: "))
Primeiro = (Numero1*2)*(Numero2/2)
Segundo = (Numero1*3)+Numero3
Terceiro = Numero3**3
print(f"""Resultado 1: {Primeiro}
Resultado 2: {Segundo}
Resultado 3: {Terceiro}""")
```

```python
#Elabore um código que receba os dados de entrada a altura de uma pessoa, e informe qual seria o peso ideal, usando a seguinte fórmula: (72.7*altura) - 58
altura = float (input("Sua altura: "))
resultado = (72.7*altura)-58
print(f"Peso ideal para sua altura: {resultado}")
```
